```python
columns = vanti.columns.to_list()
n = 1
preguntas = []
for col in columns:
    if '¿' in col:
        q = str(n) + '. ' + col
        preguntas.append(q)
        n += 1
#preguntas
```

```python
import re 
import string
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Exploración preguntas 8 y 9
print('¿En general cómo defines tu experiencia como cliente interno en Vanti?\nRespuestas:', vanti_df['¿En general cómo defines tu experiencia como cliente interno en Vanti?'].unique())

# Creación df por cada respuesta
pregunta9_df = pd.DataFrame(columns = ['01', 'respuestas_9'])
for i in vanti_df['¿En general cómo defines tu experiencia como cliente interno en Vanti?'].unique():
    respuestas = vanti_df[vanti_df['¿En general cómo defines tu experiencia como cliente interno en Vanti?']==i]['¿Cuéntanos por qué?'].str.cat(sep=' ')
    df = pd.DataFrame([[i, respuestas]], columns=['01', 'respuestas_9'])
    pregunta9_df = pregunta9_df.append(df)
```

```python
pregunta9_df.set_index('01', inplace=True)
```

```python
def clean_text1(text):
    text = text.lower()
    text = re.sub('\[.*?¿\]\%', ' ', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
 
round1 = lambda x: clean_text1(x)
 
pregunta9 = pd.DataFrame(pregunta9_df['respuestas_9'].apply(round1))
```

```python
pregunta9
```

```python
def clean_text_round2(text):
    text = re.sub('[‘’“”…«»]', '', text)
    text = re.sub('\n', ' ', text)
    return text
 
round2 = lambda x: clean_text_round2(x)
 
pregunta9 = pd.DataFrame(pregunta9['respuestas_9'].apply(round2))

vanti_df.to_pickle("corpus_pregunta9.pkl")
```

```python
pregunta9
```

```python
with open('spanish.txt') as f:
    lines = f.read().splitlines()
 
cv = CountVectorizer(stop_words=lines)
data_cv = cv.fit_transform(pregunta9['respuestas_9'])
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = pregunta9.index

data_dtm.to_pickle("dtm.pkl")

pregunta9.to_pickle('pregunta9_data_clean.pkl')
 
data_dtm
```

```python
data = pd.read_pickle('dtm.pkl')
data = data.transpose()
data
```

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(30)
    top_dict[c]= list(zip(top.index, top.values))

# Top 15 por categoria
for categoria, top_words in top_dict.items():
    print(categoria)
    print(', '.join([word for word, count in top_words[0:14]]))
```
